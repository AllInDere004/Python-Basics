### Basic program related to python

```python
name=input("Enter your name: ")
age=int(input("Enter your age: "))
print("My name is", name, "and I am", age,"years old.")  # Concatenates with commas
print(f"My name is {name} and I am {age} years old.")      # using f-string for embedding variables directly
```

```python
x=10
y="10"
print(10 + "10")    # Error: python never mixes data types automatically
print(x + int(y))  # Converts y to integer and adds
print(str(x) +y)    # Converts x to string and concatenates
```

```python
'''
int * int → numeric multiplication
str * int → string repetition (e.g., '5' * 2 → '55')
str * str → error (unsupported)
int * str → error (unsupported)  

'''
a='5'
b=2
print(a * b)  # prints '55' because string '5' is repeated 2 times

a='5'
b='2'
print(a*b)  # Error: can't multiply two strings

a=5
b='2'
print(a*b)  # Error: can't multiply integer with string
```

```python
x=3
y='2'
print(str(x) * int(y))  # Converts x to string and y to integer, then repeats string '3' two times resulting in '33'
```

```python
x=7
print(x%2)  # prints 1 because 7 divided by 2 leaves a remainder of 1
```

```python
a=5
b=2
print(a/b)  # prints 2.5 (float division)
print(a//b) # prints 2 (integer division)
```
