```python
# loop control statements
# break
# it exists the loop permaturely or permanently
for i in range(10):
    if i==7:
        break
    print(i)
```

```python
# continue
# it skip the current iteration and continues with the next
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# pass
# it is a null operation; it does nothing
for i in range(5):
    if i==3:
        print("pass it: ",i)
        pass
    print(i)
```

```python
# nested loops
# a loop inside a loop
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
# examples:- calculate the sum of N natural numbers using while and for loop
n=int(input("Enter a number: "))
i=1;sum=0
while i<n:
    sum=sum+i
    i=i+1
print("The sum of first natural number is: ",sum)

```

```python
# for loop for n natural numbers
n=int(input("Enter a number: "))
sum=0
for i in range(1,n):
    sum=sum+i
print("The sum of first natural number is: ",sum)
```

```python
# display prime numbers between 1 to 100
n=101
for i in range(1,n):
    if i>1:
        for j in range(2,i):
            if i%j==0:
                break
        else:
            print(i)
```

```python
# using while loop of prime number
count = 2
while count <= 100:
    is_prime = True
    j = 2
    while j < count:
        if count % j == 0:
            is_prime = False
            break
        j += 1
if is_prime:
        print(count)
    count += 1
```
