```python
# 23. Pascals triangle pattern
r=int(input("Enter the row size for pattern: "))
for i in range(r):
    n=1
    for j in range(r-i):
        print(" ", end=" ")
    for k in range(i+1):
        print(n, end="   ")
        n=n*(i-k)//(k+1)
    print()
```

```python
# 24. Butterfly pattern 
r=int(input("Enter the row size for pattern: "))
for i in range(1,r+1):
    for j in range(1,i+1):
        print("*", end=" ")
    for j in range(2*(r-i)):
        print(" ", end=" ")
    for j in range(1,i+1):
        print("*", end=" ")
    print()
for i in range(r,0,-1):
    for j in range(1,i+1):
        print("*", end=" ")
    for j in range(2*(r-i)):
        print(" ", end=" ")
    for j in range(1,i+1):
        print("*", end=" ")
    print()
```

```python
# 25. Hollow butterfly pattern
r=int(input("Enter the row size for pattern: "))
for i in range(1,r+1):
    for j in range(1,i+1):
        if j==1 or j==i:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    for j in range(2*(r-i)):
        print(" ", end=" ")
    for j in range(1,i+1):
        if j==1 or j==i:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
for i in range(r,0,-1):
    for j in range(1,i+1):
        if j==1 or j==i:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    for j in range(2*(r-i)):
        print(" ", end=" ")
    for j in range(1,i+1):
        if j==1 or j==i:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
# 28. Zig-zag number pattern
r=int(input("Enter the row size for pattern: "))
for i in range(1,r+1):
    for j in range(1,r+1):
        if (i+j)%2==0:
            print("1", end=" ")
        else:
            print("0", end=" ")
    print()
```

```python
# 29. Cross pattern
r=int(input("Enter the row size for pattern: "))
for i in range(1,r+1):
    for j in range(1,r+1):
        if i==j or i+j == r+1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
# 30. plus pattern 
r=int(input("Enter the row size for pattern: "))
for i in range(1,r+1):
    for j in range(1,r+1):
        if i==r//2+1 or j==r//2+1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
