```python
# Nested conditional statements
# we can put if - else in the sub group of if-else statements
# even,odd,negative

num=int(input("Enter a number: "))
if(num>0):
    print("The number is positive")
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("The number is zero or negative")
```

```python
# Practical Examples

# determine if the year is leap year or not using nested if-else statement
year=int(input("Enter the year"))
if(year%4==0):
    if(year%100==0):
        if(year%400==0):
            print(year,"is always a leap Year")
        else:
            print(year,"is not a i leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"haven't a leap year")
```
